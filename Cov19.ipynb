## Preprocessing

### Derived tables

```python
China = full_table[full_table['Country'] == 'China']
row = full_table[full_table['Country'] != 'China']
```

```python
full_latest = full_table[full_table['Date last updated'] == max(full_table['Date last updated'])].reset_index()
China_latest = full_latest[full_latest['Country'] == 'China']
row_latest = full_latest[full_latest['Country'] != 'China']
```

```python
full_latest_grouped = full_latest.groupby('Country')['Confirmed', 'Suspected', 'Recovered', 'Death'].sum().reset_index()
China_latest_grouped = China_latest.groupby('Province/State')['Confirmed', 'Suspected', 'Recovered', 'Death'].sum().reset_index()
row_latest_grouped = row_latest.groupby('Country')['Confirmed', 'Suspected', 'Recovered', 'Death'].sum().reset_index()
```

## Some Stats

### Latest data

```python
temp = full_latest.groupby('Date last updated')['Confirmed', 'Suspected', 'Recovered', 'Death'].sum().reset_index()
temp = temp.sort_values('Date last updated', ascending = False)
temp.head(1).style.background_gradient(cmap = 'Pastel1')
```

```python
temp = full_latest.groupby(['Country', 'Province/State'])['Confirmed', 'Suspected', 'Recovered', 'Death'].sum()
temp.style.background_gradient(cmap = 'Pastel1_r')
```

## Cases in different countries

```python
fig = px.choropleth(full_table, locations="Country", 
                    locationmode='country names', color="Confirmed", 
                    hover_name="Country", range_color=[1,500], 
                    color_continuous_scale="aggrnyl", 
                    title='Countries with Confirmed Cases')
fig.update(layout_coloraxis_showscale=False)
fig.show()
```

```python
fig = px.choropleth(full_table[full_table['Death']>0], 
                    locations="Country", locationmode='country names',
                    color="Death", hover_name="Country", 
                    range_color=[1,50], color_continuous_scale="agsunset",
                    title='Countries with Deaths Reported')
fig.update(layout_coloraxis_showscale=False)
fig.show()
```

```python
formated_gdf = full_table.groupby(['Date last updated', 'Country'])['Confirmed', 'Suspected', 'Recovered', 'Death'].max()
formated_gdf = formated_gdf.reset_index()
formated_gdf = formated_gdf[formated_gdf['Country'] != 'China']
formated_gdf['Date last updated'] = pd.to_datetime(formated_gdf['Date last updated'])
formated_gdf['Date last updated'] = formated_gdf['Date last updated'].dt.strftime('%Y/%m/%d')

fig = px.scatter_geo(formated_gdf[formated_gdf['Country'] != 'China'],
                     locations = "Country", locationmode = 'country names',
                     color = "Confirmed", size = 'Confirmed', hover_name = "Country",
                     range_color = [0, max(formated_gdf['Confirmed'])+2],
                     projection = "natural earth", animation_frame = "Date last updated",
                     title = 'Spread outside China over time')

fig.update(layout_coloraxis_showscale = False)
fig.show()
```

```python
China_map = China.groupby(['Date last updated', 'Province/State'])['Confirmed', 'Suspected', 'Recovered', 'Death', 
                                                      'Lat', 'Long'].max()
China_map = China_map.reset_index()
China_map['size'] = China_map['Confirmed'].pow(0.5)
China_map['Date last updated'] = pd.to_datetime(China_map['Date last updated'])
China_map['Date last updated'] = China_map['Date last updated'].dt.strftime('%Y/%m/%d')
China_map.head()

fig = px.scatter_geo(China_map, lat='Lat', lon='Long', scope='asia',
                     color="size", size='size', hover_name='Province/State', 
                     hover_data=['Confirmed', 'Suspected', 'Recovered', 'Death'],
                     projection="natural earth", animation_frame="Date last updated", 
                     title='Spread in China over time')
fig.update(layout_coloraxis_showscale=False)
fig.show()
```
